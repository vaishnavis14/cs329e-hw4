# Part III

```python
def right_ascension(hours, mins, secs):
  return (hours + (mins / 60) + (secs / 3600)) * 15
df['ra'] = right_ascension(df['RAh'], df['RAm'], df['RAs'])
df.head()
```

```python
def declination(degrees, mins, secs):
  return degrees + (mins / 60) + (secs / 3600)
df['dec'] = declination(df['DEd'], df['DEm'], df['DEs'])
df.head()
```

# Part V

```python
from astropy.coordinates import SkyCoord
from astropy import units
equatorial_coord = SkyCoord(ra=df['ra'], dec=df['dec'], unit=(units.degree, units.degree))

galactic_coord = equatorial_coord.galactic

df['galactic_longitude'] = galactic_coord.l.deg
df['galactic_latitude'] = galactic_coord.b.deg
df.head()
```

# Part VI

```python
df['color'] = np.nan
for i in range(df.shape[0]):
  letter = str(df['SpType'][i]).strip()[0].upper()
  if letter == 'O':
    df['color'][i] = 'violet'
  elif letter == 'B':
    df['color'][i] = 'mediumblue'
  elif letter == 'A':
    df['color'][i] = 'lightblue'
  elif letter == 'F':
    df['color'][i] = 'green'
  elif letter == 'G':
    df['color'][i] = 'yellow'
  elif letter == 'K':
    df['color'][i] = 'orange'
  elif letter == 'M':
    df['color'][i] = 'red'
  else:
    df['color'][i] = 'black'
```

# Part VII

```python
color_discrete_map = {'violet': 'rgb(127, 0, 255)', 'mediumblue': 'rgb(0,130,230)', 'lightblue': 'rgb(120,206,255)',
                      'green' : 'rgb(0,128,0)', 'yellow' : 'rgb(255, 204, 0)', 'orange' : 'rgb(255, 165, 0)',
                      'red' : 'rgb(139,0,0)', 'black' : 'rgb(0,0,0)'}
px.scatter(df, 'ra', 'dec', hover_name = 'Name', title = 'Equatorial Latitude vs Longitude',
           color = 'color', color_discrete_map = color_discrete_map,
           labels = {'dec' : 'Equatorial Longitude', 'ra' : 'Equatorial Latitude'})
```
